```python
import numpy as np 
import matplotlib.pyplot as plt 
def gradient_descent(X, y, learning_rate=0.01, iterations=100): 
    m = len(y) 
    n_features = X.shape[1] 
    theta = np.zeros(n_features) 
    cost_history = [] 
    for _ in range(iterations): 
        # Compute predictions 
        predictions = np.dot(X, theta) 
        # Compute error 
        error = predictions - y 
        # Compute cost (mean squared error) 
        cost = np.sum(error ** 2) / (2 * m) 
        cost_history.append(cost) 
        # Compute gradient 
        gradient = np.dot(X.T, error) / m 
        # Update parameters 
        theta -= learning_rate * gradient.sum(axis=1)  # Sum over samples 
        #theta -= learning_rate * gradient  # This can be used if each column of gradient  corresponds to one sample 
    return theta, cost_history 
# Gnerate sample data  
np.random.seed(0) 
X = 2 * np.random.rand(100, 1) 
y = 4 + 3 * X + np.random.randn(100, 1) 
# Add bias term to features 
X_b = np.c_[np.ones((100, 1)), X] 
# Run gradient descent 
theta, cost_history = gradient_descent(X_b, y) 
# Plot cost function convergence 
plt.plot(range(len(cost_history)), cost_history) 
plt.xlabel('Iterations') 
plt.ylabel('Cost') 
plt.title('Gradient Descent - Cost Function Convergence') 
plt.show() 
print("Optimal theta:", theta)
```
