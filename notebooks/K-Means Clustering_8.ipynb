```python
import numpy as np 
from sklearn.cluster import KMeans 
import matplotlib.pyplot as plt 
# Data points 
data = np.array([[1, 2], [1, 4], [1, 0],[10, 2], [10, 4], [10, 0]]) 
# Create and fit the model 
kmeans = KMeans(n_clusters=2, random_state=0).fit(data) 
# Predict the cluster for new data points 
new_data = np.array([[0, 0], [12, 3]]) 
predictions = kmeans.predict(new_data) 
# Plotting the results 
plt.scatter(data[:, 0], data[:, 1], c=kmeans.labels_, cmap='viridis', marker='o', label='Training data') 
plt.scatter(new_data[:, 0], new_data[:, 1], c=predictions, cmap='viridis', marker='s', label='New data') 
plt.title('Data Clustering with K-Means') 
plt.xlabel('Feature 1') 
plt.ylabel('Feature 2') 
plt.legend() 
plt.show() 
print(f"Cluster+ predictions for new data: {predictions}")
```
